```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading dataset
churn_data = pd.read_csv('Churn_Modelling.csv')
```

```python
churn_data.head()
```

```python
# Splitting the dataset into target and feature
X = churn_data.iloc[:, 3:13]
y = churn_data.iloc[:, 13]
```

```python
# Feature Engineering - Handling the categorical values
Geography_enc = pd.get_dummies(X['Geography'],drop_first=True,dtype = int)
Gender_enc = pd.get_dummies(X['Gender'],drop_first=True,dtype = int)
#Dropping the original columns
X = X.drop(['Geography','Gender'],axis=1)
```

```python
X = pd.concat([X,Geography_enc,Gender_enc],axis=1)
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
