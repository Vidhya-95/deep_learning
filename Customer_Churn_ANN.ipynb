```python
import tensorflow as tf
print(tf.__version__)
```

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading dataset
churn_data = pd.read_csv('Churn_Modelling.csv')
```

```python
churn_data.head()
```

```python
# Splitting the dataset into target and feature
X = churn_data.iloc[:, 3:13]
y = churn_data.iloc[:, 13]
```

```python
# Feature Engineering - Handling the categorical values
Geography_enc = pd.get_dummies(X['Geography'],drop_first=True,dtype = int)
Gender_enc = pd.get_dummies(X['Gender'],drop_first=True,dtype = int)
#Dropping the original columns
X = X.drop(['Geography','Gender'],axis=1)
```

```python
X = pd.concat([X,Geography_enc,Gender_enc],axis=1)
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.layers import ReLU, LeakyReLU, PReLU, ELU
```

```python
#Initializing the model
ann_model = Sequential()
```

```python
#Creating Input Layer and first hidden layer (both are same) - We have 11 input parameters which will be expected by the hidden layer
ann_model.add(Dense(units=8,activation='relu',input_dim=11))
```

```python
# Creating second hidden layer
ann_model.add(Dense(units=10,activation='relu'))
```

```python
# Creating output layer - Since we are having binary target variable , we will have only one neuron in the output layer
ann_model.add(Dense(units=1,activation='sigmoid'))
```

```python
ann_model.summary()
```

```python
# Compiling the model
# Here we have a binary classification , therefore we will be using binary_crossentropy as our loss function.
# Optimizer is adam and metric to be evaluated is accuracy

ann_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=10,
    restore_best_weights=True
)
```

```python
# Train the model
churn_train_history = ann_model.fit(X_train,y_train, batch_size =100,epochs=200,callbacks=[early_stopping],validation_split=0.2)
```
